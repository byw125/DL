Word cloud 보기

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

user_text = input("Please paste your text here: ")

wordcloud = WordCloud(
    background_color='white',
    width=800,
    height=800,
    max_words=100
).generate(user_text)

plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
%%capture
!pip install gtts
from gtts import gTTS
from IPython.display import Audio, display

def tts(text):
  gtts_object = gTTS(text,lang = "en", slow = False)
  gtts_object.save("myaudio.mp3")
  return Audio("myaudio.mp3")
```

```python
!pip install pyqrcode

import pyqrcode
from pyqrcode import QRCode

s = input("Paste the link: ")

url = pyqrcode.create(s)

url.svg("myqrcode.svg", scale=10)

from IPython.display import SVG, display
def show_svg(file):
    display(SVG(file))

show_svg("myqrcode.svg")
```
